```python
# import libraries
import pandas as pd
```

Read the csv file created in the project-gnod notebook

```python
all_w_audio_ft = pd.read_csv('all_w_audio_ft.csv')
all_w_audio_ft
```

```python
all_w_audio_ft_num = all_w_audio_ft.select_dtypes(exclude='object')
all_w_audio_ft_cat = all_w_audio_ft.select_dtypes('object')
```

```python
display(all_w_audio_ft_num)
display(all_w_audio_ft_cat)
```

```python
# check if 'id' and 'song_id' columns contain the same information
list(all_w_audio_ft_cat['id'])==list(all_w_audio_ft_cat['song_id'])
```

```python
# drop 'id' column
all_w_audio_ft_cat = all_w_audio_ft_cat.drop(['id'], axis=1)
```

```python
all_w_audio_ft_cat.to_csv('all_w_audio_ft_cat.csv', sep=',', index=False)
```

```python
# the scale of "popularity" is much higher than the scale of many other variables!
# K-Means is a distance based algorithm: we need to scale / normalize:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(all_w_audio_ft_num)
X_scaled = scaler.transform(all_w_audio_ft_num)

# now, all features will have the same weight
pd.DataFrame(X_scaled,columns=all_w_audio_ft_num.columns).head()
```

```python
# pickle scaler and X_scaled
import pickle

pickle.dump(scaler, open('scaler.p', 'wb'))
pd.DataFrame(X_scaled).to_pickle('X_scaled.p')
```
